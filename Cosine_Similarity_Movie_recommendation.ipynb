Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Read Data

```python
df = pd.read_csv("/content/movie_dataset.csv")
```

```python
df.head()
```

generate features

```python
features = ['keywords', 'cast', 'genres', 'director']
```

```python
for feature in features:
  df[feature] = df[feature].fillna('')
```

```python
def combined_features(row):
  return row['keywords']+" " +row['cast']+" "+row['genres']+" "+row['director']
df['combined_features'] = df.apply(combined_features,axis=1)
```

**Extracting Features**


sklearn.feature_extraction module can be used to extract features in format supported by machine learning algorithms


CountVectorizer()'s *fit_transform* will help to count the number of text

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combined_features'])
print("Count Matrix: ",count_matrix.toarray())
```

**Cosine Similarity**

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
movie_user_like ="Dead Poets Society"

def get_index_from(title):
  # Filter the DataFrame for the given title
  movie_row = df[df.title == title]
  # Check if the movie was found
  if not movie_row.empty:
    return movie_row["index"].values[0]
  else:
    # Handle the case where the movie is not found, e.g., print an error and return None
    print(f"Error: Movie '{title}' not found in the dataset.")
    return None

movie_index = get_index_from(movie_user_like)
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

**Generating similar movies matrix**

```python
print(similar_movies)
```

```python
sorted_similar_movies = sorted(similar_movies, key = lambda x:x[1],reverse= True)
```

```python
print(sorted_similar_movies)
```

**Printing the Similar Movies**

```python
def get_title_from_index(index):
  return df[df.index == index]["title"].values[0]

i = 0
for movies in sorted_similar_movies:
  print(get_title_from_index(movies[0]))
  i=i+1;
  if i>15:
    break
```
